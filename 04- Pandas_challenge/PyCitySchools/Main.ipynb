# School Summary

```python
BySchool_name = school_data_complete.set_index("school_name").groupby(["school_name"])
School_type = schools_df.set_index("school_name")["type"]

Total_studentper_school = BySchool_name["Student ID"].count()
Total_school_budget = schools_df.set_index("school_name")["budget"]
Per_student_budget = schools_df.set_index("school_name")["budget"]/schools_df.set_index("school_name")["size"]

Average_Math = school_data_complete["math_score"].mean()
Average_Reading = school_data_complete["reading_score"].mean()
Pass_math = school_data_complete[school_data_complete['math_score']>=70].groupby("school_name")["Student ID"].count()/Total_studentper_school
Pass_reading = school_data_complete[school_data_complete['reading_score']>=70].groupby("school_name")["Student ID"].count()/Total_studentper_school
Overall_Pass = school_data_complete[(school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70)].groupby("school_name")["Student ID"].count()/Total_studentper_school

School_summary = pd.DataFrame({
                "School type":School_type,
                "Total Students":Total_studentper_school,
                "Total School Budget": Total_school_budget,
                "Per Student Budget": Per_student_budget,
                "Average Math Score":Average_Math,
                "Average Reading Score":Average_Reading,
                "% Passing Math": Pass_math,
                "% Passing Reading":Pass_reading,
                "% Overall Passing":Overall_Pass })

School_summary = School_summary[[
                   "School type",
                    "Total Students",
                    "Total School Budget",
                     "Per Student Budget",
                    "Average Math Score",
                    "Average Reading Score",
                    "% Passing Math",
                    "% Passing Reading",
                     "% Overall Passing"]]

School_summary.head(5).style.format({
                     "Total Students": "{:,.0f}",
                    "Total School Budget": "${:,.0f}",
                     "Per Student Budget": "${:,.0f}",
                    "Average Math Score": "{:,.2f}",
                    "Average Reading Score":"{:,.2f}",
                    "% Passing Math": "{:.1%} ",
                    "% Passing Reading": "{:.1%}",
                     "% Overall Passing":"{:.1%}"})
                         
```

# Top Performing Schools (By % Overall Passing)

```python
Top_performance = School_summary.sort_values("% Overall Passing", ascending = False)

Top_performance.head().style.format({
                    "Total Students": "{:,}",
                    "Total School Budget": "${:,.0f}",
                     "Per Student Budget": "${:,.0f}",
                    "Average Math Score": "{:,.2f}",
                    "Average Reading Score":"{:,.2f}",
                    "% Passing Math": "{:.1%} ",
                    "% Passing Reading": "{:.1%}",
                     "% Overall Passing":"{:.1%}"})
 
```

# Math Scores by Grade

# Scores by School Spending

```python

bins = [0, 584.9999, 629.9999, 644.9999, 680]
group = ["<585", "585-630", "630-645", "645-680"]
school_data_complete["bin_group"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"],bins,labels=group)
group_spending=school_data_complete.groupby("bin_group")

average_math = group_spending["math_score"].mean()
average_read = group_spending["reading_score"].mean()
passing_math = school_data_complete[school_data_complete["math_score"]>=70].groupby("bin_group")["Student ID"].count()/group_spending["Student ID"].count()
passing_reading = school_data_complete[school_data_complete["reading_score"]>=70].groupby("bin_group")["Student ID"].count()/group_spending["Student ID"].count()
Overall_passing=school_data_complete[(school_data_complete["reading_score"]>=70) & (school_data_complete["math_score"]>=70)].groupby("bin_group")["Student ID"].count()/group_spending["Student ID"].count()

Spending_Score_df= pd.DataFrame({
                        "Average Math Score":average_math,
                         "Average Reading Score":average_read,
                        "% Passing Math":passing_math,
                        "% Passing Reading": passing_reading,
                        "% Overall Passing": Overall_passing })

Spending_Score_df.index.name = "Spending ranges per student($)"

Spending_Score_df.style.format({
                        "Average Math Score": "{:.2f}",
                        "Average Reading Score": "{:.2f}",
                        "% Passing Math":"{:.1%}",
                        "% Passing Reading":"{:.1%}",
                        "% Overall Passing": "{:.1%}"})
```

# Scores by School Size

```python
bins = [0, 999.99, 1999.99, 5000]
group_size = ["Small < 1000", "Medium (1000-2000)", "Large (2000-5000)"]

school_data_complete["school_size"] = pd.cut(school_data_complete["size"], bins,labels=group_size)
school_size_group = school_data_complete.groupby("school_size")

average_math_size = school_size_group["math_score"].mean()
average_read_size = school_size_group["reading_score"].mean()
passing_math_size = school_data_complete[school_data_complete["math_score"]>=70].groupby("school_size")["Student ID"].count()/school_size_group["Student ID"].count()
passing_reading_size = school_data_complete[school_data_complete["reading_score"]>=70].groupby("school_size")["Student ID"].count()/school_size_group["Student ID"].count()
Overall_passing_size = school_data_complete[(school_data_complete["reading_score"]>=70)&(school_data_complete["math_score"]>=70)].groupby("school_size")["Student ID"].count()/school_size_group["Student ID"].count()

Overall_sore_by_size = pd.DataFrame ({
            "Average Math Score": average_math_size,
            "Average Reading Score": average_read_size,
            "% Passing Math": passing_math_size,
            "% Passing Reading": passing_reading_size,
            "% Overall Passing": Overall_passing_size})

Overall_sore_by_size.index.name = "School by size" 

Overall_sore_by_size.style.format({
            "Average Math Score": "{:.2f}",
            "Average Reading Score": "{:.2f}",
            "% Passing Math": "{:.1%}",
            "% Passing Reading": "{:.1%}",
            "% Overall Passing":"{:.1%}"})
```

# Scores by School Type

```python
school_type = school_data_complete.groupby("type")

average_math_type = school_type["math_score"].mean()
average_reading_type = school_type["reading_score"].mean()
passing_math_type= school_data_complete[school_data_complete["math_score"]>=70].groupby("type")["Student ID"].count()/school_type["type"].count()
passing_reading_type = school_data_complete[school_data_complete["reading_score"]>=70].groupby("type")["Student ID"].count()/school_type["type"].count()
overall_passing_type = school_data_complete[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)].groupby("type")["Student ID"].count()/school_type["type"].count() 

Over_score_bytype = pd.DataFrame({
                "Average Math Score":average_math_type,
                "Average Reading Score": average_reading_type,
                "% Passing Math":passing_math_type,
                "% Passing Reading":passing_reading_type ,
                "% Overall Passing":overall_passing_type})

Over_score_bytype.index.name= "School type"
        
Over_score_bytype.style.format({
                "Average Math Score": "{:.1f}",
                "Average Reading Score": "{:.1f}",
                "% Passing Math": "{:.1%}",
                "% Passing Reading": "{:.1%}",
                "% Overall Passing": "{:.1%}"})
```
